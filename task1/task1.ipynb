```python
import numpy as np 
import pandas as pd
```

```python
df=pd.read_csv("KaggleV2-May-2016.csv")
```

```python
df.describe()
```

```python
#Checking details of the dataframe for clear understanding of requirements
df.info()
```

```python
# Check for null values
df.isnull().sum()
```

```python
# Removing duplicates
df = df.drop_duplicates()
```

```python
# Standardizing values
#Gender
if 'Gender' in df.columns:
    df['Gender'] = df['Gender'].str.upper().str.strip()

#Dates
for col in ['ScheduledDay', 'AppointmentDay']:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col])
```

```python
#Rename columns 
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
#convert datatype of age to Integer
if 'age' in df.columns:
    df['age'] = pd.to_numeric(df['age']).astype('Int64')
#convert datatype of age to Integer
if 'neighbourhood' in df.columns:
    df['neighbourhood'] = df['neighbourhood'].astype(str).str.strip().str.title()
```
